```python
x= np.linspace(-2,2,50)
def sigmoid(x):
    return 1/(1+np.exp(-x))


def swish(x):
    return x*(1/(1+np.exp(-x)))

plt.plot(x,sigmoid(x), 'g+',label = 'sigmoid function')
plt.plot(x,swish(x),'b^', label = 'swish function')


plt.xlabel('x values')
plt.ylabel('y values')

plt.title('Sigmoid and Swish graph')
plt.legend()
plt.show()
```

```python
#plot a decaying function: half life of C14 isotope using matplotlib

# half life = tou * (ln(2))

# N(t) = N0(1/2)(t/t2)  t - elapsed time , t2 - half life

N_start = 100
half_life = 1.1

def half_life_cal(N_start, time, half_life):
    return N_start * (np.exp(-half_life*time))

time = np.arange(0,15,1)

plt.plot(time,half_life_cal(N_start, time, half_life))
plt.title("Half life")
plt.show()
```

```python
# To Design an Email Filter
'''You are given an integer followed by email addresses. Your task is to print a list containing only valid email addresses in lexicographical order.


Valid email addresses must follow these rules:

It must have the username@websitename.extension
The username can only contain letters, digits, dashes and underscores .
The website name can only have letters and digits .
The extension can only contain letters .
The maximum length of the extension is 3
Input Format
The first line of input is the integer , the number of email addresses.
lines follow, each containing a string.

Constraints

Each line is a non-empty string.

Sample Input

3
lara@golang.com
brian-23@firebird.com
britts_54@nocode4#u.com
'''
import re
allowed_username = re.compile('[a-zA-Z0-9_-]+$')
allowed_websitename = re.compile('[a-zA-Z0-9]+$')
allowed_extension = re.compile('[a-zA-Z]+$')
                       
num_email = int(input("Enter number of emails: ")) 
email_list= []
for i in range(num_email):
    print("Enter email number", i+1, ": " )
    item = input()
    email_list.append(item)
    
#username  = set(email_list[0].split('@'))
#print(username[0])

for item in email_list:
    username =  email_list[0].split('@')
    if allowed_username.match(username[0]):
        print("\n The username of", item, "is Valid")                       
    else:
        print("The username of", item, "is Invalid")
        
    websitename = username[1].split('.')
    if allowed_websitename.match(websitename[0]):
        print("The websitename of", item, "is Valid")
    else:
        print("The websitename of", item, "is Invalid")
                          
    #extension = username[1].split('.')
    if allowed_extension.match(websitename[1]) and len(websitename[1]) < 4:
        print("The extension of",item,"is Valid")
    else:
        print("The extension of", item, "is Invalid")
        print("\n")
```
