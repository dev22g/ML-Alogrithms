```python
'''

Covariance:Measures the relationship btween two random variables. This metric Evaluates to what extent the variables change together. Essentially its a measure of variance between 2 variables.

positive covariance - ( x and Y) together go up
negative covariance - x and y together go down

Correlation: direction and strength
scaled version of covariance

corr = Cov(x,y)/std(x)*std(y)

pearson product moment correlation cofficient (rho)
x and y  - linear relationship
'''
```

```python
import numpy as np
```

```python
#100 random integers
x = np.random.randint(0,50,100)

#positive relationship
# adding to random noise
y = x +np.random.normal(0,10,100)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
```

```python
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
np.cov(x,y) # covariance matrix
```

```python
np.corrcoef(x,y)
```
